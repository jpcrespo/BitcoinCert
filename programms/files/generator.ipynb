```python
import os
from fpdf import FPDF
from faker import Faker

fake = Faker()

def create(batch):
    folder_name = os.path.join("batch", str(batch))
    os.makedirs(folder_name, exist_ok=True)

    for _ in range(batch):

        name = fake.name()
        cert_id = fake.uuid4()
        

        pdf = FPDF(orientation="L", unit="mm", format="A4")
        pdf.add_page()
        

        pdf.image("logo.png", x=-10, y=2, w=100)    
        pdf.set_line_width(2)
        pdf.rect(10, 10, 277, 190) 
        
        pdf.set_font("helvetica", "B", 25)
        pdf.text(x=70, y=30, txt="King Fahd University of Petroleum & Minerals")
        pdf.text(x=70, y=40, txt="College of Computing and Matematics")
        pdf.text(x=70, y=50, txt="Information and Computer Science Department")

        pdf.set_font("helvetica", "B", 45)
        pdf.text(x=80, y=80, txt="Academic Diploma")
        
        pdf.set_font("helvetica", "", 20)
        pdf.text(x=115, y=90, txt="Awarded to:")
        
        pdf.set_font("helvetica", "I", 35)
        pdf.text(x=100, y=115, txt=name)
        
        pdf.set_font("helvetica", "I", 15)
        pdf.text(x=80, y=125, txt="For successfully completing the course for Bitcoiners in 2026.")
        
        pdf.set_font("helvetica", "", 12)
        pdf.text(x=105, y=180, txt=f"ID de Verificación: {cert_id}")
        namefile = f"{name.replace(' ', '_')}_{cert_id[:6]}.pdf"
        path = os.path.join(folder_name, namefile) 
        
        pdf.output(path)


create(1)
create(5)
create(10)
```
