```python
import matplotlib.pyplot as plt

N_batch = [1, 8, 16, 32, 64, 128, 256, 512, 1024, 1984]
grow_proof = [543, 743, 810, 877, 944, 1011, 1078, 1145, 1212, 1279]
mean_verify_time = [0.0078, 0.0122, 0.0149, 0.018, 0.0225, 0.022, 0.024, 0.0291, 0.0309, 0.0337]
std_ver_time = [0.0036, 0.0022, 0.0025, 0.0043, 0.0054, 0.0039, 0.0025, 0.0063, 0.0068, 0.0075]
revocation_vsize = [170, 170, 171, 173, 177, 185, 201, 233, 298, 420]
revocation_size = [221, 221, 222, 224, 228, 236, 252, 284, 349, 471]
```

```python
plt.semilogx(N_batch, grow_proof, marker='o', linestyle='-', color='b')
plt.xlabel('Batch Size ($N\_batch$)')
plt.ylabel('GROW ($grow\_proof$ [bytes])')
plt.title('GROW vs Batch Size')
plt.grid(True)
plt.tight_layout()
plt.savefig('grow_vs_batch_size.png')
```

```python

plt.semilogx(N_batch, mean_verify_time, marker='o', linestyle='-', color='r')
plt.xlabel('Batch Size ($N\_batch$)')
plt.ylabel('Verify Time (sec)')
plt.title('Verify Time vs Batch Size')
plt.grid(True)
plt.tight_layout()
plt.savefig('verify_time_vs_batch_size.png')
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

# Graficar
plt.semilogx(N_batch, revocation_vsize, marker='o', label='Virtual Size (vbytes)')
plt.semilogx(N_batch, revocation_size, marker='s', label='Total Size (bytes)')

plt.xlabel('Batch Size ($N\_batch$)')
plt.ylabel('Size')
plt.title('Size Revocation Transaction vs Batch Size')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.5)

# Caption
plt.figtext(0.5, 0.01, "", 
            ha="center", fontsize=10, bbox={"facecolor":"orange", "alpha":0.2, "pad":5})

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.savefig('revocation_size_vs_batch.png')
plt.show()
```
